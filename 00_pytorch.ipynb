Two main ways of multiplying matrix in deep learning and neural networks
1) Element wise mul.
2) Matrix Multiplication through dot product.                       
~There are two rule that performing mat mul need to satisfy otherwise we see errors
The inner dimension must match.     
'(3,2) @ (3,2)' this would not work.  
'(3,2) @ (2,3)' this would work.  
~The result has the shape of outer dimension


```python
torch.matmul(tensor,tensor)
```

```python
%%time
value = 0
for i in range(len(tensor)):
  value+=tensor[i]*tensor[i]
print(value)
```

## Reshaping, viewing and stacking Tensors
Reshaping : reshape an input tensor into defined shape      
View: Return the view of an input tensor of a certain shape but keeping the same mem as the original tensor.  
Stacking: Combining all tensor on top of eachother(vstack) or side by side(hstack)
Squeezing: Removing 1 D tensors from stack.  
unsqueezing: Adding 1 D tensors from stack.  
Permute: Return a view of input with dimensions permuted(swapped) in a certain way

```python
tensor = torch.arange(0,12)
tensor.shape
```

```python
reshaping_tensor = torch.reshape(tensor,(3,4))
reshaping_tensor.shape
reshaping_tensor
```

```python
#view
tensor.view(2,6)
```

```python
x_staked = torch.stack([tensor,tensor,tensor],dim=0)
x_staked
#
```

```python
tensor.unsqueeze(dim=0)
```

```python
tensor.squeeze()
```
