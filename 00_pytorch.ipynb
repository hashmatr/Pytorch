## **Tensor datatypes**

```python
float_32_tensor = torch.tensor([3.0,4.0,5.0])
float_32_tensor
```

Convert from float 32 to float 16

```python
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
float_32_tensor*float_16_tensor
```

Tensors not right datatype - to do get datatype from a tenso can use 'tensor.dtype'

Tensors not on the right device - We can use tensor.device
